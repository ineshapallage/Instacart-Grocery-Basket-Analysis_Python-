## Q6 Create a spending flag for each user based on the average price across all their orders using the following criteria:.”

    -If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
    -If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

##### groups dataframe by “user_id” column &e transform() function on the “prices” column to generate the average prict'

```python
ords_prods_merge['average_spending']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

##### Assign spending flags

```python
ords_prods_merge.loc[ords_prods_merge['average_spending']>=10,'spending flag']='High spender'
ords_prods_merge.loc[ords_prods_merge['average_spending']<10,'spending flag']='Low spender'
```

```python
ords_prods_merge[['user_id','loyalty_flag', 'average_spending', 'spending flag']].head(50)
```

## Q7:  Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:r.”

    -If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

    -If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

    -If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Frequent customer or Not

#Group dataframe by 'user_id', calculate median of 'days_since_prior_order' and assign flags

ords_prods_merge['frequency_of_purchase']=ords_prods_merge.groupby(['user_id'])['days_from_last_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['frequency_of_purchase']>20,'order frequency flag']='Non-frequent customer'

ords_prods_merge.loc[(ords_prods_merge['frequency_of_purchase']>10) & (ords_prods_merge['frequency_of_purchase']<=20),'order frequency flag']='Regular customer'

ords_prods_merge.loc[ords_prods_merge['frequency_of_purchase']<=10,'order frequency flag']='Frequent customer'
```

```python
ords_prods_merge['order frequency flag'].value_counts(dropna=False)
```
